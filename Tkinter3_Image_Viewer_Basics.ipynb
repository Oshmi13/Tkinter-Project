Using Icons, Images, and Exit Buttons

```python
from tkinter import *


root = Tk()
root.title('Tasveer')

# .ico file has to be downloaded. It is necessari in tkinter window
root.iconbitmap('D:/Images/Icon_files/File.svg')

button_quit = Button(root, text="Exit Program", command = root.destroy)
button_quit.pack()

root.mainloop()
```

```python
from tkinter import *
from PIL import ImageTk, Image

root = Tk()
root.title('Tasveer')
# .ico file has to be downloaded. It is necessari in tkinter window
root.iconbitmap('D:/Images/Icon_files/File.svg')

# Step 1 - define image
my_img = ImageTk.PhotoImage(Image.open("D:/Images/Icon_files/Nezuko.jpg"))
# Step 2 - Store it somewhere
my_label = Label(image=my_img)
# Step 3 - Pack it on the screen
my_label.pack()

button_quit = Button(root, text="Exit Program", command = root.destroy)
button_quit.pack()

root.mainloop()
```
