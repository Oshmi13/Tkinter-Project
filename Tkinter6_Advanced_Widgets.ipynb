RadioButtons

```python
from tkinter import*
from PIL import ImageTk,Image

root =Tk()
root.title('HELLO')
root.iconbitmap('D:/Images/Icon_files/File.svg')  

r = IntVar()
r.set("-1")
# variable is always called using get 

def clicked(value):
    myLabel = Label(root, text=value)
    myLabel.pack()

Radiobutton(root, text ="Options 1", variable = r, value=1, command = lambda: clicked(r.get())).pack()
Radiobutton(root, text ="Options 2", variable = r, value=2, command = lambda: clicked(r.get())).pack()

myLabel = Label(root, text=r.get())
myLabel.pack()

myButton = Button(root, text="Click Me!", command = lambda: clicked(r.get()))
myButton.pack()

root.mainloop()
```

```python
from tkinter import*
from PIL import ImageTk,Image

root =Tk()
root.title('HELLO')
root.iconbitmap('D:/Images/Icon_files/File.svg')  


TOPICS = [ 
    # ("TEXT","MODE")
    ("COA","GATE1"),
    ("OS", "GATE2"),
    ("DBMS","GATE3"),
    ("TOC","GATE4"),
]

future = StringVar()
future.set("JUST DO IT!")

for text, mode in TOPICS:
    Radiobutton(root, text=text, variable=future, value=mode).pack(anchor = W)

def clicked(value):
    myLabel = Label(root, text=value)
    myLabel.pack()


myLabel = Label(root, text=future.get())
myLabel.pack()

myButton = Button(root, text="Click Me!", command = lambda: clicked(future.get()))
myButton.pack()

root.mainloop()
```

Message Box

```python
from tkinter import*
from PIL import ImageTk,Image
from tkinter import messagebox

root =Tk()
root.title('HELLO')
root.iconbitmap('D:/Images/Icon_files/File.svg')  

# showinfo, showwarning, showerror, askquestion, askokcancel, askyesno
def popup1():
   messagebox.showinfo("This is my popup!","Hello world!")
def popup2():
   messagebox.showwarning("This is my popup!","Wrong Password")
def popup3():
   messagebox.showerror("This is my popup!","Hello world!")
def popup4():
   messagebox.askquestion("This is my popup!","Hello world!")

#askokcancel # askyesno

B_1= Button(root, text = "Info!", command=popup1).pack(anchor = W)
B_2= Button(root, text = "Alert!", command=popup2).pack(anchor = W)
B_3= Button(root, text = "Error!", command=popup3).pack(anchor = W)
B_4= Button(root, text = "Sure!", command=popup4).pack(anchor = W)


mainloop()
```

```python
from tkinter import*
from PIL import ImageTk,Image
from tkinter import messagebox

root =Tk()
root.title('HELLO')
root.iconbitmap('D:/Images/Icon_files/File.svg')  

# showinfo, showwarning, showerror, askquestion, askokcancel, askyesno
def popup():
    response = messagebox.askquestion("This is my popup!","Hello world!")
    Label(root, text=response).pack()
    if response =="yes":
            Label(root, text="You Clicked me!").pack()
    else:
            Label(root, text ="You Clicked No").pack()

B = Button(root, text = "Sure!", command=popup).pack()


mainloop()
```
