```python
from tkinter import *

root = Tk()

# Creating Label widget
# First Define the label
# Then display/pack it on the screen 
myLabel = Label(root, text ="Hello World!")

#Shoving it onto the screen
myLabel.pack()

root.mainloop()

```

Positioning with Tkinter's Grid System

```python
from tkinter import *

root = Tk()

# Creating Label widget
# First Define/Create the label
# Then display/pack it on the screen 
myLabel1 = Label(root, text ="Hello World!")
myLabel2 = Label(root, text ="             ")
myLabel3 = Label(root, text ="I am KalluKalia")

myLabel1.grid(row=0, column=0)
myLabel2.grid(row=1, column=1)
myLabel3.grid(row=1, column=10)

# myLabel2.grid(row=1, column=5) with column 5 no evident result is observed
# Becoz Tkinter ignores the command becoz their is no label in col 123

root.mainloop()
```

Creating Buttons

```python
from tkinter import *

root = Tk()

def myClick():
    myLabel = Label(root, text ="Look! I clicked a Button!")
    myLabel.pack()

myButton1 = Button(root, text="Click Me!", padx=50, pady =50, command = myClick, fg="white", bg="blue")
myButton2 = Button(root, text="Do not click Me!", state=DISABLED, command=myClick)
# Using command we have to call the function for Label to execute
# do not use brackets while calling a function like myClick() while creating buttons
# fg = changing colour for button's text 
# bg = changing button's colour
# can use hex HTML colour codes 
myButton1.pack()
myButton2.pack()


root.mainloop()
```

Creating Input Fields

```python
from tkinter import *

root = Tk()

e = Entry(root, width = 50, borderwidth = 5)
# can use fg and bg in Entry
e.pack()

def myClick():
    myLabel = Label(root, text ="Look! I clicked a Button!")
    myLabel.pack()

myButton = Button(root, text="Click Me!", command = myClick)

myButton.pack()

root.mainloop()
```

get Function

```python
from tkinter import *

root = Tk()

e = Entry(root, width = 50, borderwidth = 5)
# can use fg and bg in Entry
e.pack()


def myClick():
    myLabel = Label(root, text =e.get())
    myLabel.pack()

myButton = Button(root, text="Enter Your Name", command = myClick)

myButton.pack()

root.mainloop()
```

```python
from tkinter import *

root = Tk()

e = Entry(root, width = 50, borderwidth = 5)
# can use fg and bg in Entry
e.pack()
e.insert(0, "Enter Your name: ")


def myClick():
    hello ="Hello  " + e.get()
    myLabel = Label(root, text = hello)
    myLabel.pack()

myButton = Button(root, text="Enter Your Name", command = myClick)

myButton.pack()

root.mainloop()
```
