Build An Image Viewer App

```python
from tkinter import *
from PIL import ImageTk, Image

root = Tk()
root.title('Tasveer')
root.iconbitmap('D:/Images/Icon_files/File.svg')


my_img1 = ImageTk.PhotoImage(Image.open("D:/Images/Icon_files/Whatsapp_Image.jpg"))
my_img2 = ImageTk.PhotoImage(Image.open("D:/Images/Icon_files/Whatsapp_Image2.jpg"))
my_img3 = ImageTk.PhotoImage(Image.open("D:/Images/Icon_files/Whatsapp_Image3.jpg"))
my_img4 = ImageTk.PhotoImage(Image.open("D:/Images/Icon_files/Whatsapp_Image4.jpg"))
#my_img4 = ImageTk.PhotoImage(Image.open("D:/Images/Icon_files/Todolist.jpg"))
#my_img5 = ImageTk.PhotoImage(Image.open("D:/Images/Icon_files/Whatsapp_Image5.jpg"))

image_list = [my_img1, my_img2, my_img3, my_img4]

my_label = Label(image=my_img1)
my_label.grid(row=0, column=0, columnspan=3)

def forward(image_number):
    global my_label
    global button_forward
    global button_back
    
    # forget the current image
    my_label.grid_forget()
    # Pack the next new image
    my_label = Label(image=image_list[image_number-1])
    button_forward = Button(root, text=">>", command=lambda: forward(image_number+1))
    button_back = Button(root, text="<<", command=lambda: backward(image_number-1))
    
    if image_number == 4:
        button_forward = Button(root, text=">>", state=DISABLED)

    my_label.grid(row=0, column=0, columnspan=3)
    button_back.grid(row=1, column=0)
    button_forward.grid(row=1, column=2)

def backward(image_number):
    global my_label
    global button_forward
    global button_back

    my_label.grid_forget()
    my_label = Label(image=image_list[image_number-1])
    button_forward = Button(root, text=">>", command=lambda: forward(image_number+1))
    button_back = Button(root, text="<<", command=lambda: backward(image_number-1))
    
    if image_number == 1:
        button_backward = Button(root, text="<<", state=DISABLED)
    
    my_label.grid(row=0, column=0, columnspan=3)
    button_back.grid(row=1, column=0)
    button_forward.grid(row=1, column=2)

    
    

button_back = Button(root, text="<<", command=lambda: backward(1), state=DISABLED)
button_exit = Button(root, text="Exit Program", command = root.destroy)
button_forward = Button(root, text=">>", command=lambda: forward(2))

button_back.grid(row=1, column=0)
button_exit.grid(row=1, column=1)
button_forward.grid(row=1, column=2)



root.mainloop()
```
