Open file dialog box

```python
from tkinter import*
from PIL import ImageTk,Image
from tkinter import filedialog

root =Tk()
root.title('HELLO')
root.iconbitmap('D:/Images/Icon_files/File.svg')

def open():
    global my_image
    root.filename = filedialog.askopenfilename(initialdir=" D:/Images/Icon_files", title = "Select a file", filetypes=(("png files","*.png"), ("all files","*.*")))
    # We will pack the image location and the images uisng Label on the tk window 
    my_label = Label(root, text=root.filename).pack()
    my_image = ImageTk.PhotoImage(Image.open(root.filename))
    my_image_label = Label(image=my_image).pack()

my_btn = Button(root, text="Open file", command= open).pack()


root.mainloop()
```

```python
from tkinter import*
from PIL import ImageTk,Image

root =Tk()
root.title('HELLO')
root.iconbitmap('D:/Images/Icon_files/File.svg')
root.geometry("400x400")

vertical =Scale(root, from_=0, to=200)
vertical.pack()

def slide(var):
    my_label = Label(root, text=horizontal.get()).pack()
    root.geometry(str(horizontal.get()) + "x400")

horizontal =Scale(root, from_=0, to=400, orient=HORIZONTAL, command = slide)
horizontal.pack()


my_btn = Button(root, text="Click Me!", command = slide).pack()


root.mainloop()
```
